## Using the tab option for VEP output

```python
import pandas as pd
import gzip
vcf_file = "/home/dmc2245/test/bcftools/snps_indels_vep_test_sorted.VEP.CADD_gnomAD.tsv.gz"
```

```python
df = pd.read_csv(vcf_file, comment='#', sep='\t', header=None)
```

```python
df
```

```python
# Initialize an empty list to store the comments
comments = []

# Open the gzip-compressed VCF file and read it line by line
with gzip.open(vcf_file, 'rt') as file:
    for line in file:
        if line.startswith('#'):
            comments.append(line.strip())
```

```python
# The last item in the 'comments' list is the header comment
header_comment = comments[-1]
```

```python
header_comment
```

```python
# Extract the header from the comment
header = header_comment.split('#')[1]
```

```python
header
```

```python
# Split the header into column names
columns = header.split('\t')
```

```python
columns
```

```python
df = pd.read_csv(vcf_file, comment='#', sep='\t', names=columns, compression='gzip')
```

```python
df
```

```python
df[['Uploaded_variation','Existing_variation']]
```

```python
# Split the values in 'your_column' and create new columns
df[['chromosome', 'position']] = df['Location'].str.split(':', expand=True)
```

```python
df
```

```python
df[['start', 'end']] = df['position'].str.split('-', expand=True)
```

```python
df
```

```python
# If there are only two values, set the end column equal to the start column
df['end'].fillna(df['start'], inplace=True)
```

```python
df
```

```python
# Drop the intermediate columns
df.drop(['position'], axis=1, inplace=True)
```

```python
column_order = ['chromosome', 'start', 'end'] + [col for col in df.columns if col not in ['chromosome', 'start', 'end']]
df = df[column_order]
```

```python
df
```

```python
# Print the result
print(df)
```

```python
df.to_csv('/home/dmc2245/test/bcftools/test_out.tsv', index=False, sep='\t', header=True)
```
