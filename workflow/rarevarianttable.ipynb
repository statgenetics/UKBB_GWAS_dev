````sos
def transform_exome(statsfile,outfile):
    import pandas as pd
    import numpy as np
    data = pd.read_csv(statsfile,header=0,sep="\t",compression="gzip")
    out = data[["CHROM","GENPOS","ALLELE1","ALLELE0","BETA","SE","LOG10P"]]
    out["LOG10P"] = 10**(-out["LOG10P"])
    out.columns = ["chromosome","base_pair_location","effect_allele","other_allele","beta","standard_error","p_value"]
    out = out.replace(to_replace =np.nan, value ="NA")
    out.to_csv(outfile,index=False,sep="\t")
````
