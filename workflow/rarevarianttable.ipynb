### MASK1

```sos
raw = pd.read_csv("~/UKBiobank/results/REGENIE_results/results_burden_exome/012822_anno_tables/f2247.GJB2.Mask1.0.01/f2247.GJB2.Mask1.0.01.raw",sep=" ",header=0)
raw
```

```sos
raw["sum"] = raw.iloc[:,6:].sum(axis=1,skipna=True)
raw["sum"].value_counts()
```

```sos
cols = raw[raw["sum"]==2].sum(axis=0,skipna=True)
raw[raw["sum"]==2][[cols[cols!=0].index][0]]
```

```sos
raw["carrier"] = 0
raw.loc[raw["sum"] > 0,"carrier"] = 1
raw["carrier"].value_counts()
```

```sos
carrierlist = raw[raw["carrier"]==1]["IID"].to_list()
phe[phe["IID"].isin(carrierlist)]
```

```sos
noncarrierlist = raw[raw["carrier"]==0]["IID"].to_list()
```

```sos
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (5, 5))
rare_variants = ['carrier', 'noncarrier']
c = [phe.loc[phe["IID"].isin(carrierlist),"age"].std(),phe.loc[phe["IID"].isin(noncarrierlist),"age"].std()]
average_age = [phe.loc[phe["IID"].isin(carrierlist),"age"].mean(), phe.loc[phe["IID"].isin(noncarrierlist),"age"].mean()]
plt.bar(rare_variants, average_age, color ='maroon',width = 0.3)
plt.errorbar(rare_variants, average_age, yerr=c, fmt="o", color="black")
plt.xlabel("Rare Variants")
plt.ylabel("Average Age")
plt.show()
```

```sos
average_age
```

```sos
c
```

```sos
from scipy.stats import ttest_ind
ttest_ind(phe.loc[phe["IID"].isin(carrierlist),"age"],phe.loc[phe["IID"].isin(noncarrierlist),"age"])
```
